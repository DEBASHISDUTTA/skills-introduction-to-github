```python
import pandas as pd

# Load the Excel file
file_path = '/content/Walmart_store_data.xlsx'
xls = pd.ExcelFile(file_path)

# Check sheet names to select appropriate one
sheet_names = xls.sheet_names
sheet_names
```

```python
# Load the data from 'Sheet1'
data = pd.read_excel(file_path, sheet_name='Sheet1')

# Display the first 5 rows
data.head()
```

```python
import matplotlib.pyplot as plt

# Ensure 'Store_ID' and 'Week' are treated appropriately
unique_stores = data['Store_ID'].unique()

# Set up the plot
plt.figure(figsize=(14, 8))

# Plot each store's revenue trend
for store_id in unique_stores:
    store_data = data[data['Store_ID'] == store_id]
    plt.plot(store_data['Week'], store_data['Weekly_Revenue'], label=store_id)

# Enhance the visualization
plt.title('Revenue Trends vs Week for All Stores', fontsize=16)
plt.xlabel('Week', fontsize=14)
plt.ylabel('Weekly Revenue', fontsize=14)
plt.legend(title='Store ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Analyze the correlation between 'Weekly_Revenue' and 'Weekly_Customer_Visits'
correlation = data['Weekly_Revenue'].corr(data['Weekly_Customer_Visits'])

correlation
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_excel('Walmart_store_data.xlsx', sheet_name='Sheet1')

# Group data by 'Weather' and calculate average revenue
weather_revenue = data.groupby('Weather')['Weekly_Revenue'].mean().sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(12, 7))
weather_revenue.plot(kind='bar', color='skyblue', edgecolor='black')

# Enhance the chart with professional labeling
plt.title('Average Weekly Revenue by Weather Conditions', fontsize=18)
plt.xlabel('Weather Condition', fontsize=14)
plt.ylabel('Average Weekly Revenue', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the bar chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel('Walmart_store_data.xlsx', sheet_name='Sheet1')

# Group by Store_ID and calculate average turnover rate
store_turnover = data.groupby('Store_ID')['Turnover_Rate'].mean().sort_values(ascending=False)

# Set up the figure
plt.figure(figsize=(14, 8))

# Create a bar chart
store_turnover.plot(kind='bar', color='mediumseagreen', edgecolor='black')

# Professional chart labeling
plt.title('Average Turnover Rate by Store', fontsize=18)
plt.xlabel('Store ID', fontsize=14)
plt.ylabel('Average Turnover Rate (%)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Highlight the highest and lowest turnover stores
highest_store = store_turnover.idxmax()
lowest_store = store_turnover.idxmin()

# Annotate the highest turnover rate
plt.annotate(f'Highest: Store {highest_store}',
             xy=(store_turnover.index.get_loc(highest_store), store_turnover.max()),
             xytext=(store_turnover.index.get_loc(highest_store)-1, store_turnover.max()+0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=12, color='black')

# Annotate the lowest turnover rate
plt.annotate(f'Lowest: Store {lowest_store}',
             xy=(store_turnover.index.get_loc(lowest_store), store_turnover.min()),
             xytext=(store_turnover.index.get_loc(lowest_store)+1, store_turnover.min()-1),
             arrowprops=dict(facecolor='red', arrowstyle='->'),
             fontsize=12, color='red')

# Display the chart
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel('Walmart_store_data.xlsx', sheet_name='Sheet1')

# Select numerical columns only
numerical_features = data.select_dtypes(include='number')

# Generate correlation matrix
correlation_matrix = numerical_features.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Feature Correlation Matrix', fontsize=18)
plt.show()
```
